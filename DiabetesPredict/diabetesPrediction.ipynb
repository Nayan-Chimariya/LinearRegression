**Imports**

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

**Reading Data**

```python
train_data = pd.read_csv("training_data.csv")
test_data = pd.read_csv("testing_data.csv")
```

**Extract the features and lables**

```python
train_data.head()
```

```python
feature_columns = train_data.drop(columns=['Outcome']).columns.tolist()
train_features = train_data[feature_columns].values
train_labels = train_data['Outcome'].values

feature_columns = test_data.drop(columns=['Outcome']).columns.tolist()
test_features = test_data[feature_columns].values
test_labels = test_data['Outcome'].values
```

```python
model = keras.Sequential([
    layers.Input(shape=(train_features.shape[1],)),
    layers.Dense(1)
])
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae','accuracy'])
```

```python
model.fit(train_features, train_labels, epochs=100, verbose=1)
```

```python
eval_loss, eval_mae, eval_accuracy = model.evaluate(test_features, test_labels)
print(f"Evaluation Mean Absolute Error: {eval_mae}")
print(f"Accuracy: {eval_accuracy}")
```
